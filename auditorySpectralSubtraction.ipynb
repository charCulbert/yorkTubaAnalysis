```python
SAMPLE_RATE = 22050
yamaha, sr = librosa.load('AUDIO_DEMOS/sm57/tuned_scales/yamaha/combinationMethodTuned_yamaha_sm57_chromaticScale.wav', sr=SAMPLE_RATE)
york, sr = librosa.load('AUDIO_DEMOS/sm57/tuned_scales/york/combinationMethodTuned_york_sm57_chromaticScale.wav', sr=SAMPLE_RATE)
```

```python
D_york = np.fft.fft(york)
D_yamaha = np.fft.fft(yamaha)

mag_york = abs(D_york)
phase_york = np.angle(D_york)

mag_yamaha = abs(D_yamaha)
phase_yamaha = np.angle(D_yamaha)

mag_subtractedYorkMinusYamaha = mag_york - mag_yamaha
mag_subtractedYorkMinusYamaha = np.where(mag_subtractedYorkMinusYamaha < 0, 0, mag_subtractedYorkMinusYamaha)
reconstructedComplexYorkMinusYamaha = mag_subtractedYorkMinusYamaha * np.exp(1j * phase_york)

mag_subtractedYamahaMinusYork = mag_yamaha - mag_york
mag_subtractedYamahaMinusYork = np.where(mag_subtractedYamahaMinusYork < 0, 0, mag_subtractedYamahaMinusYork)
reconstructedComplexYamahaMinusYork = mag_subtractedYamahaMinusYork * np.exp(1j * phase_yamaha)

reconstructedYamahaMinusYork = np.real(np.fft.ifft(reconstructedComplexYamahaMinusYork))
reconstructedYorkMinusYamaha = np.real(np.fft.ifft(reconstructedComplexYorkMinusYamaha))

print('yamaha')
# librosa.display.waveshow(yamaha, sr=sr)
print('york')
# librosa.display.waveshow(york, sr=sr)



print("Yamaha - York")
display(Audio(reconstructedYamahaMinusYork, rate=sr))
# librosa.display.waveshow(reconstructedYamahaMinusYork, sr=sr)

print("York - Yamaha")
display(Audio(reconstructedYorkMinusYamaha, rate=sr))
# librosa.display.waveshow(reconstructedYorkMinusYamaha, sr=sr)
```
